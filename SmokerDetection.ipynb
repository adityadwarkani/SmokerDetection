# Importing training data from folder

```python
import os
from PIL import Image
import pandas as pd

def resize_image(image, size):
    return image.resize(size)

folder_path = '/Users/adi/Desktop/archive/Training/Training'

image_data = []
image_metadata = []

desired_size = (250, 250)

for file_name in os.listdir(folder_path):
    if file_name.endswith('.jpg'): 
        file_path = os.path.join(folder_path, file_name)
        try:
            image = Image.open(file_path)
            # Resize the image
            resized_image = resize_image(image, desired_size)
            # Get image data (pixels)
            data = list(resized_image.getdata())
            if file_name.startswith('notsmoking'):
                new = 0
                data.append(new)
            else:
                new = 1
                data.append(new)
            # Get image metadata
            metadata = {
                'file_name': file_name,
                'width': resized_image.width,
                'height': resized_image.height,
                'format': resized_image.format,
            }

            image_data.append(data)
            image_metadata.append(metadata)
        except Exception as e:
            print("Error processing file:", file_path)
            print(e)

df_data_train = pd.DataFrame(image_data)
df_metadata_train = pd.DataFrame(image_metadata)
```

# Importing testing data from folder


```python
import os
from PIL import Image
import pandas as pd

def resize_image(image, size):
    return image.resize(size)

folder_path = '/Users/adi/Desktop/archive/Testing/Testing'

image_data = []
image_metadata = []

desired_size = (250, 250) 

for file_name in os.listdir(folder_path):
    if file_name.endswith('.jpg'): 
        file_path = os.path.join(folder_path, file_name)
        try:
            image = Image.open(file_path)
            # Resize the image
            resized_image = resize_image(image, desired_size)
            # Get image data (pixels)
            data = list(resized_image.getdata())
            if file_name.startswith('notsmoking'):
                new = 0
                data.append(new)
            else:
                new = 1
                data.append(new)
            # Get image metadata
            metadata = {
                'file_name': file_name,
                'width': resized_image.width,
                'height': resized_image.height,
                'format': resized_image.format,
            }

            image_data.append(data)
            image_metadata.append(metadata)
        except Exception as e:
            print("Error processing file:", file_path)
            print(e)

df_data_test = pd.DataFrame(image_data)
df_metadata_test = pd.DataFrame(image_metadata)
```

# Extracting features

```python
X = df_data_train.iloc[:,:-1]
X
```

```python
Y=df_data_train.iloc[:,-1]
Y
```

```python
testing_X = df_data_test.iloc[:,:-1]
testing_X
```

```python
testing_Y=df_data_test.iloc[:,-1]
testing_Y
```

# Flattening the tuples to put into model

```python
flattened_columns = {}
for column in X.columns:
    flattened_columns.update({
        f'{column}_element_{i}': [t[i] for t in X[column]] for i in range(len(X[column][0]))
    })

flattened_df = pd.DataFrame(flattened_columns)

print(flattened_df)
```

```python
flattened_columns = {}
for column in testing_X.columns:
    flattened_columns.update({
        f'{column}_element_{i}': [t[i] for t in testing_X[column]] for i in range(len(testing_X[column][0]))
    })

flattened_df_testing = pd.DataFrame(flattened_columns)

print(flattened_df_testing)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report
```

# Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
model.fit(flattened_df,Y)
```

```python
y_pred=model.predict(flattened_df_testing)
y_pred
```

```python
accuracy=accuracy_score(testing_Y,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(testing_Y,y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", annot_kws={"size": 16})
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
plt.title("Confusion Matrix")
plt.show()
print(classification_report(testing_Y,y_pred))
```
